```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gennorm
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "arial",
    "font.size": 31})
```

```python
def gen(beta):
    x = np.linspace(gennorm.ppf(0.01, beta),gennorm.ppf(0.99, beta), 100)
    return x
```

```python
beta1 = 1
beta2 =2 
beta3 =3
curt_list = [beta1, beta2, beta3]
f = plt.figure(figsize = (12,3))
plt.style.use('ggplot')
for N, beta in enumerate(curt_list):
    pos = 130+(N+1)
    a = f.add_subplot(pos)
    x = gen(beta)
    a.plot(x, gennorm.pdf(x, beta), color = 'r')
    if beta == 1:
        plt.title('Leptocúrtica', fontsize = 27)
    if beta == 2:
        plt.title('Mesocúrtica', fontsize = 27)
    if beta == 3:
        plt.title('Platicúrtica', fontsize = 27)
    plt.grid(True)
    plt.xticks([])
    plt.yticks([])
    #plt.axis('off')
#plt.figure(linewidth=10, edgecolor = '#04253a')
plt.savefig('curtose.pdf')
```
