```python
import matplotlib.pyplot as plt
import numpy as np
import random
import seaborn as sns
import scipy.stats as st
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "arial",
    "font.size": 31})
```

```python
def sorteio1(n):
    soma = 0
    somas = list()
    for x in range(0,n):
        a = random.randint(1,12)
        b = random.randint(1,12)
        c = random.randint(1,12)
        soma = a+b+c
        somas.append(soma)
    return somas
```

```python
s5 = sorteio1(5)
s51 = sorteio1(50)
s52 = sorteio1(500)
s53 = sorteio1(5000)
s54 = sorteio1(50000)
s55 = sorteio1(500000)
```

```python
import scipy.stats as st
s_list = [s5, s51, s52, s53, s54, s55]
letras = ['(a)','(b)','(c)','(d)', '(e)', '(f)']
f = plt.figure(figsize = (19,15))
plt.style.use('ggplot')
for N, sm in enumerate(s_list):
    mu, sigma = np.mean(sm), np.var(sm)
    pos = 230+(N+1)
    a = f.add_subplot(pos)
    if len(sm) >= 5000:
        b = 32
        r = 0.6
    else:
        b = 12
        r = 0.6
    n,x,_ = a.hist(sm,density = True,bins = b, alpha = 0.75, rwidth = r, color = 'cornflowerblue')
    density = st.gaussian_kde(sm)
    plt.grid(axis = 'y', alpha = 0.75)
    plt.title(f'{letras[N]} Número de amostras: ${len(sm)}$', fontsize=28)
    plt.xlabel('Somas', fontsize = 36)
    plt.xticks(fontsize = 26)
    plt.yticks(fontsize = 26)
    plt.subplots_adjust(wspace = 0.45, hspace = 0.3)
    plt.ylabel('Densidade', fontsize = 36)
    plt.plot(x, density(x), linestyle = 'dashed', lw = 2, color = 'red')
plt.savefig('histograma9.pdf')
```
