```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
from matplotlib import pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "arial",
    "font.size": 24})
```

```python
plt.figure(figsize = (19,12))
v = venn2(subsets = (1, 1, 1), set_labels = ('$H(X)$', '$H(Y)$'), set_colors=('b', 'r'), alpha = 0.8);
venn2_circles(subsets = (1,1,1));
v.get_patch_by_id("10").set_color('cornflowerblue')
v.get_label_by_id("10").set_text("$H(X|Y)$")
v.get_patch_by_id("11").set_color("lightsteelblue")
v.get_label_by_id("11").set_text("$I(X;Y)$")
v.get_label_by_id("01").set_text("$H(Y|X)$")
v.get_patch_by_id("01").set_color("darksalmon")
plt.title('$H(X,Y)$')
plt.savefig('venn_mi2.pdf')
```
